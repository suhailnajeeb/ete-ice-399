# Experiment 3: Amplitude Modulation (AM)

```
import numpy as np
from numpy import cos, absolute
from numpy.fft import fft,fftshift
from math import pi
import matplotlib.pyplot as plt
```

## DSB-FC

```
fs = 8000
fc = 100
Ac = 1
fm = 10
Am = 0.6
```

```
t = np.arange(0,1,1/fs)
M = Am*cos(2*pi*fm*t)
C = Ac*cos(2*pi*fc*t)
S_AM = (Ac+M)*cos(2*pi*fc*t)
```

```
plt.plot(t,M)
plt.xlabel('Time (s)')
plt.ylabel('m(t)')
plt.xlim((0,5/fm))
plt.title('Message Signal in Time Domain')
plt.show()
```

```
plt.plot(t,C)
plt.xlabel('Time (s)')
plt.ylabel('c(t)')
plt.xlim((0,5/fm))
plt.title('Carrier Signal in Time Domain')
plt.show()
```

```
plt.plot(t,S_AM)
plt.xlabel('Time (s)')
plt.ylabel('S_{AM}(t)')
plt.xlim((0,5/fm))
plt.title('DSB+C Signal in Time Domain')
plt.show()
```

```
M_f = fftshift(absolute(fft(M)))
f = np.linspace(-fs/2, fs/2, len(M_f))
```

```
plt.stem(f, M_f/M_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |M(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of Message Signal')
plt.show()
```

```
C_f = fftshift(absolute(fft(C)))
```

```
plt.stem(f, C_f/C_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |C(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of Carrier Signal')
plt.show()
```

```
S_AM_f = fftshift(absolute(fft(S_AM)))
```

```
plt.stem(f, S_AM_f/S_AM_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |S_{AM}(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of DSB+C Signal')
plt.show()
```
